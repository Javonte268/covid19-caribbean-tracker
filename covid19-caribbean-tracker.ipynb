```python
covid_data = pd.read_csv('/Users/javonteanthony/Documents/covid.csv',header=1)
```

```python
df = pd.DataFrame(covid_data)
```

```python
caribbean_countries = [
    'Anguilla', 'Antigua and Barbuda', 'the British Virgin Islands', 'Guadeloupe', 'Montserrat','Saint Kitts and Nevis', 'Saint Martin','the United States Virgin Islands'
]
```

```python
caribbean_df = df[df['Entity'].isin(caribbean_countries)].copy()
```

```python
caribbean_df = caribbean_df.dropna(subset=['Total confirmed deaths due to COVID-19'])
caribbean_df = caribbean_df[caribbean_df['Total confirmed deaths due to COVID-19'] > 0]
```

```python
caribbean_df['Day'] = pd.to_datetime(caribbean_df['Day'])
```

```python
plt.figure(figsize=(14, 7))

for country in caribbean_countries:
    country_df = caribbean_df[caribbean_df['Entity'] == country]
    country_df = country_df.sort_values('Day')
    plt.plot(country_df['Day'], country_df['Total confirmed deaths due to COVID-19'], label=country)

plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Total Confirmed Deaths')
plt.title('COVID-19 Confirmed Deaths in Caribbean Countries')
plt.legend(loc='upper left', fontsize='small', ncol=2)
plt.tight_layout()
plt.grid(True)
```
